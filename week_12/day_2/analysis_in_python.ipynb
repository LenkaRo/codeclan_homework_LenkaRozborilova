5. Summarise the mean number of calories in each beverage_category. Which 3 category of drinks have the most calories? Which 3 drinks have the least? Write a small summary of your findings.

6. Harder - Which Beverage_prep type contains more than the average calories of all drinks? (Hint: to answer this, you’ll have to first figure out what the average calories across all drinks are, and then use that as a filter for the grouped Beverage_prep data).

```python
drinks
```

```python
# average calories accross all drinks
avg_cal = drinks["Calories"].mean()
avg_cal
```

```python
# Which Beverage_prep type contains more than the average calories of all drinks?

# create new column `average_within_Beverage_prep_group` and populate it with average of calories of the group
drinks["average_within_Beverage_prep_group"] = drinks.groupby("Beverage_prep")["Calories"].transform("mean")

# add logical operator that compares the groups calories average with the overall calories average
drinks["more_than_overall_average"] = drinks["average_within_Beverage_prep_group"] > avg_cal
```

## Extension - Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
drinks_content = pd.read_csv("data/starbucks_drinkMenu_expanded.csv")
drinks_content.head()
```

```python
drinks_content_plot_data = drinks_content.groupby("Beverage_category")["Calories"].mean().reset_index(name = "avg_cal").sort_values("avg_cal")
drinks_content_plot_data
```

```python
sns.barplot(x = "avg_cal",
           y = "Beverage_category",
           data = drinks_content_plot_data);
plt.ylabel(" ")
plt.title("Average amount of calories in each of Starbucks beverage category");
```
